```python
def rating_category_function(rating):

    if rating >= 7.6:
        return "High"
    elif rating >= 5.6:
        return "Medium"
    else:
        return "Low"
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg,sum, desc

spark = SparkSession.builder.appName("MoviesSet").getOrCreate()
```

```python
Spdf = spark.createDataFrame(movieRecord)
```

```python
Spdf.show()
```

```python
# Group by runtimeMinutes and genres, then calculate the average rating for each group
runtime_genre_ratings = Spdf.groupBy("runtimeMinutes", "genres").agg(avg("averageRating").alias("avg_rating"))

# Display the results, sorted by average rating in descending order
runtime_genre_ratings.orderBy(desc("avg_rating")).show()
```

```python
# Which movie genre has the highest average rating?
# Group by genres and calculate the average rating for each genre
genre_ratings = Spdf.groupBy("genres").agg(avg("averageRating").alias("avg_rating"))

# Find the genre with the highest average rating
highest_rated_genre = genre_ratings.orderBy(desc("avg_rating")).limit(1)

# Display the result
highest_rated_genre.show()
```

```python
# QUESTION 6. What is the average number of votes for movies released in each region?
# Group by region and calculate the average number of votes
region_votes = Spdf.groupBy("region").agg(avg("numVotes").alias("avg_votes"))

# Order the results by average votes in descending order
region_votes.orderBy(desc("avg_votes")).show()
```

```python
#Which primary professions of contributors have the highest average movie ratings?
# Group by primary profession and calculate the average rating
profession_ratings = Spdf.groupBy("primaryProfession").agg(avg("averageRating").alias("avg_rating"))

# Order the results by average rating in descending order
profession_ratings.orderBy(desc("avg_rating")).show()
```

```python
# what is the average number of votes for movie released in each region?
year_ratings = Spdf.groupBy("startYear").agg(avg("averageRating").alias("avg_rating"))
year_ratings.orderBy("startYear").show()
```

```python
# List top 5 most frequent genre produce each year
def mapper(record):
    year = record['startYear']
    genres = record['genres'].split(',')
    for genre in genres:
        yield (year, genre), 1
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, explode, split, desc
spark = SparkSession.builder.appName("MovieGenresByYear").getOrCreate()
Spdf = spark.createDataFrame(movieRecord)
genres_by_year = Spdf.select("startYear", explode(split(col("genres"), ",")).alias("genre"))
genre_counts = genres_by_year.groupBy("startYear", "genre").count()
top_genres = genre_counts.orderBy("startYear", desc("count")).limit(5)
top_genres.show()
```
